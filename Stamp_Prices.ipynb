```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

Dataset citation: https://vincentarelbundock.github.io/Rdatasets/datasets.html

```python
# Load Dataset
data = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/Stat2Data/USstamps.csv')
```

```python
# Create a copy for safety purposes
df = data.copy()
```

```python
df.head()
```

```python
# Plot the year vs price

#Plot titles
plt.title("Prices Over Time (US)")
plt.xlabel("Years")
plt.ylabel("Prices")

# Set columns to plot
plt.scatter(df['Year'], df['Price']) # (X axis, Y axis)
plt.show()
```
